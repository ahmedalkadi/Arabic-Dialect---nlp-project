```python

# Send a POST request to the API
response = requests.post(url, json=data)
result = response.json()

# Print the result
if response.status_code == 200:
        # Get the first 5 keys and values from the dictionary
        countries = {k: result[k] for k in list(result.keys())[:5]}

        # Sort the first 5 probabilities by probability in descending order
        sort_countries = sorted(countries.items(), key=lambda x: x[1], reverse=True)

        # Print the sorted first 5 probabilities
        print("Countries dialect probability:")
        for country, probability in sort_countries:
        # Convert probability to float if it's not the predicted country
                if isinstance(probability, float):
                        probability_str = f"{probability:.1f}%"
                else:
                        probability_str = probability
                print(f"{country}: {probability_str}")

        # Print the predicted country dialect
        predicted_country = result['predicted_country']
        print("\nThe predicted country dialect is:", predicted_country)

else:
        print("Error:", response.json())
```
